```python


## Preparing the plots
fig_hist, ax_hist = plt.subplots(10,1,figsize=(20,20))
fig_hist.subplots_adjust(top=8, bottom=4)

## Filling the plots with histograms and boxplots
for i in range(len(train_data.columns)):
    noNanData = train_data.iloc[:,i]
    noNanData = noNanData.fillna("")
    noNanData.value_counts().sort_values(axis='index').plot(kind='bar', ax=ax_hist[i])

    ax_hist[i].set_title('Histogram of attribute "{0}"'.format(train_data.columns[i]))
```

## Convert each nominal value to a boolean column

```python
def processNominalColumn(dictionary, values, columnPrefix):
    uniqueValues = values.unique()
    print('The unique values in the column are: {0}'.format(uniqueValues))
    
    for uniqueValue in uniqueValues:
        if isinstance(uniqueValue, str):
            columnName = '{0}{1}'.format(columnPrefix, uniqueValue.replace(" ", ""));
        else:
            columnName = '{0}{1}'.format(columnPrefix, uniqueValue)

        dict[columnName] = np.where(values.str.startswith(uniqueValue),  1,0)

        if np.unique(dict[columnName]).size == 1:
            dictionary.pop(columnName)
        else:
            print('Column {0} filled with {1} ....'.format(columnName, dict[columnName][0:10,]))
    print(processedDataset.keys())
    return;
```

```python
print('Processing column "{0}"'.format(train_data.columns[6]))

uniqueValues = train_data['SexuponOutcome'].unique()
print('The unique values in the column are: {0}'.format(uniqueValues))
for uniqueValue in uniqueValues:
    if isinstance(uniqueValue, str):
        columnName = 'sexIs{0}'.format(uniqueValue.replace(" ", ""));
    else:
        columnName = 'sexIsNan'
    
    processedDataset[columnName] = np.where(train_data['SexuponOutcome'].str.startswith(uniqueValue),  1,0)
    
    if np.unique(processedDataset[columnName]).size == 1:
        processedDataset.pop(columnName)
    else:
        print('Column {0} filled with {1} ....'.format(columnName, processedDataset[columnName][0:10,]))

processedDataset['isSpayedFemale'] = train_data['SexuponOutcome']

print(processedDataset.keys())
```

```python
print('Processing column "{0}"'.format(train_data.columns[6]))

processNominalColumn(processedDataset, train_data.columns['SexuponOutcome'], 'sexIs')
```
